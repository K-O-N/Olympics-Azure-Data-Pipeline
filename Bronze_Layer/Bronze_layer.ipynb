## Parameterize data for all source data

```python
dbutils.widgets.text("file_name", "")
```

```python
pfile_name = dbutils.widgets.get("file_name")
```

```python
# Stream Data using Auto Loader

df = spark.readStream.format("cloudFiles")\
          .option("cloudFiles.format", "csv")\
          .option("cloudFiles.schemaLocation", f"abfss://bronze@olympiastorage.dfs.core.windows.net/checkpoint_{pfile_name}")\
          .load(f"abfss://olympia-data@olympiastorage.dfs.core.windows.net/{pfile_name}")   
```

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
    .option("checkpointLocation", f"abfss://bronze@olympiastorage.dfs.core.windows.net/checkpoint_{pfile_name}")\
    .option("path", f"abfss://bronze@olympiastorage.dfs.core.windows.net/{pfile_name}")\
    .trigger(once=True)\
    .start()
```
